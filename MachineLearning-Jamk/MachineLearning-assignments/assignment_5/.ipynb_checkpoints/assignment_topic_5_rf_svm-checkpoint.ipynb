### Topic: Data Preprocessing (2 points)
The dataset has a class variable (*target*) in the last column.

Let's do the following preprocessing steps with the dataset:

1. Load the data into the `DataFrame`.

   * Optimize the data input to the machine learning model

```python
# TODO: Write your code in this cell
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

df = pd.read_excel('data/Date_Fruit_Datasets.xlsx')
df_plot = df.drop(columns = ['Class'])
```

```python
# TODO: Plotting features 
scaler = MinMaxScaler()

# Normalizing the data, because their scale is different.
# Otherwise hard to visualize all of them together
# Without scaling, no conclusion could be drawn from the plot.

df_scaled = scaler.fit_transform(df_plot)
df_scaled = pd.DataFrame(df_scaled, columns=df_plot.columns)

# Create a box plot to see the distribution of each feature
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_scaled, orient='h')
plt.title('Box Plot of Features')
plt.xlabel('Values')
plt.ylabel('Features')
plt.grid(True)
plt.show()
```

```python
# Investigating feature importances
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split

X = df_scaled  # Features
y = df['Class']  # Target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Create a Random Forest classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model to your data
model_rf = rf.fit(X, y)

# Get feature importances
importances = model_rf.feature_importances_

permutation_importances = permutation_importance(model_rf, X_test, y_test, n_repeats=10, random_state=42)

# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange feature names so they match the sorted feature importances
names = [X.columns[i] for i in indices]

sorted_idx = permutation_importances.importances_mean.argsort()[::-1]

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.title("Feature Importances default")
plt.bar(range(X.shape[1]), importances[indices], align="center")
plt.xticks(range(X.shape[1]), names, rotation='vertical')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.show()

# Permutation Importances
plt.figure(figsize=(10, 6))
plt.bar(df_scaled.columns[sorted_idx], permutation_importances.importances_mean[sorted_idx])
plt.xlabel('Permutation Importance')
plt.title('Permutation Importances (test set)')
plt.xticks(rotation=90)
plt.show()
```

## Analysis
Some of the features have a large range and many outliers. Having outliers does not mean data is bad.
We do not have much information what the features are. Hard to make educated guesses on the data.

### Default Feature Importance vs Permutation Importance

Default feature importance calculation is computedfrom Random Forest structure.It is based on mean decrease impurity (Gini Importance?). It measures contribution of  each featureto the reduction in impurity in decision tree forest.

Permutation feature importance, introduces shuffling of each feature and its impact on model's performance. It is based on how much model's performance drops when values of particular feature and target are shuffled. Features that cause greatest drop are considered important. This method is considered more reliable when features have non-linear relationship with target or when interactions of features are complex.

### Conclusion
The results are drastically different. Needs further testing.
The data will be tested with using both results from two different feature importances forcomparison

## Task 5.3: Implement Support Vector Machine Classification (2 points)

1. Use the `scikit-learn` library's *Linear Support Vector Classification* function to train a classifier with your training data that classifies each date fruit species.

2. Create predictions for the data points of your test data in the `y_pred` variable.

3. Use the `metrics` module to make a [classification report](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html) for the predictions of your test data. Save results to the `cr` variable.

4. Use the `metrics` module to make a **confusion matrix** and store it in the `cm` variable.

```python
# TODO: Write your implementation in this cell.
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# TODO: use LinearSVC.
svm_classifier_full = SVC(kernel='rbf')
svm_classifier_selected = SVC(kernel='rbf')

# Using two different data frames
full_features = svm_classifier_full.fit(X_trainff, y_trainff)
selected_features = svm_classifier_selected.fit(X_trainS, y_trainS)

y_pred_full = svm_classifier_full.predict(X_testff)
y_pred_selected = svm_classifier_selected.predict(X_testS)

# Classification report for the predictions
cr_full = classification_report(y_testff, y_pred_full)
cr_selected = classification_report(y_testS, y_pred_selected)

# Confusion matrices
cm_full = confusion_matrix(y_testff, y_pred_full)
cm_selected = confusion_matrix(y_testS, y_pred_selected)
```

```python
# Answers to the assignment. Note! Do not edit this cell, just run it after you complete the task.
# The results are in cm and cr variables.
print(f'Classification report (Full features):\n {cr_full}')
print(f'Classification report (Selected features):\n {cr_selected}')
print(f'Confusion matrix (Full features):\n {cm_full}')
print(f'Confusion matrix (Selected features):\n {cm_selected}')
```

```python
# Plot the confusion matrices
class_names_full = full_features.classes_
plt.figure(figsize=(8, 6))
sns.heatmap(cm_full, annot=True, cmap='Blues', fmt='g', cbar=False, xticklabels=class_names_full, yticklabels=class_names_full)
plt.title('Confusion Matrix (Full Features)')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

class_names_selected = selected_features.classes_
plt.figure(figsize=(8, 6))
sns.heatmap(cm_selected, annot=True, cmap='Blues', fmt='g', cbar=False, xticklabels=class_names_selected, yticklabels=class_names_selected)
plt.title('Confusion Matrix (Selected Features)')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```
