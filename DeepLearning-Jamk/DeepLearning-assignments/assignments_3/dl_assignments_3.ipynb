````python
# TODO: Implementation of the assignment
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout, concatenate
from tensorflow.keras.models import Model

# TODO: define three input layer
input1 = Input(shape=(10,), name = 'input_1')
input2 = Input(shape=(10,), name = 'input_2')
input3 = Input(shape=(10,),name = 'input_3')

# TODO: Middle layer
dense1_input1 = Dense(32, activation='relu', name = 'dense1_input1')(input1)
dense2_input1 = Dense(16, activation='relu', name = 'dense2_input1')(dense1_input1)
dense3_input1 = Dense(8, activation='relu', name = 'dense3_input1')(dense2_input1)

# TODO: left branch
# Define dense layers for input 2
dense1_input2 = Dense(16, activation='relu', name = 'dense1_input2')(input2)
concatenated_input2 = concatenate([dense1_input2, dense3_input1], axis = 1)
dropout_input2 = Dropout(0.2, name = 'left_dropout')(concatenated_input2)
dense2_input2 = Dense(16, activation='relu', name = 'dense2_input2')(dropout_input2)

# TODO: right branch
# Define dense layers for input 3
dense1_input3 = Dense(16, activation='relu', name = 'dense1_input3')(input3)
concatenated_input3 = concatenate([dense3_input1, dense1_input3], axis = 1)
dropout_input3 = Dropout(0.3, name = 'right_dropout')(concatenated_input3)
dense2_input3 = Dense(16, activation='relu', name = 'dense2_input3')(dropout_input3)

# TODO: join the left and right branches to the Add layer
add = tf.keras.layers.Add(name='add')([dense2_input2, dense2_input3])
final_dense1 = Dense(16, activation='relu', name='final_dense1')(add)
final_dense2 = Dense(1, activation='softmax', name='final_dense2')(final_dense1)

# TODO: Set the model you created in the model_functional variable.
# model_functional = tf.keras.Model(inputs= ..., outputs=...)
model_functional = Model(inputs=[input1, input2, input3], outputs=final_dense2, name='model_functional')
````

### Topic: Convolutional neural network and feature matrices (4 points)

**Background**

`Conv2D` layers transform the input image into a mathematical model that is used by tightly coupled layers to create a classification.
Since `Dense` layers can only handle 1D data, first the multidimensional feature map given by the last `Conv2D` layer must be converted to 1D format using the `Flatten` layer.

The following information is required for model configuration:

* the batch size is set to 50. These 50 samples are fed to the model at each step.
* Images are 32×32 pixel RGB images and are set to (32, 32, 3)
* Loss function is `sparse_categorical_crossentropy` and `optimizer` is `Adam()`
* There are 10 different classes in the data.
* The number of epoch cycles is only five (5) iterations. In real apps, there may be 50-100 rounds, but in training tasks, additional rounds are not more useful.
* 20% of the data is test data.
* 80% of the data is training data.

### Input data preprocessing

* Before fitting the neural network, convert the data to `float32` format, resulting in a faster training process on GPU processing.
* Data scaling ensures that the weight updates are smaller, which improves the final output and eases data processing.

### Building a deep learning model

When using this layer as the first layer in a model, you should provide the keyword argument `input_shape`.
In this example, these inputs are 28x28 RGB images with `channels_last`:

```>>> input_shape = (28, 28, 3)

```

1. In the task, two `Conv2D` layers are specified in consecutive order (3×3 kernel size + ReLU activation),
    and the numbers of filters for them are 32 and 64.
2. Next, we use the `Flatten` layer to convert the data into a 1D format.
3. There are two `Dense` layers (number of filters 64 and 10) to generate the classification.
   The last layer is not activated with ReLU, but with Softmax, resulting in a multi-class probability distribution in response to the question
   "which class is most likely?".

#### Compiling and fitting the model

After defining the deep learning architecture, we can build the model and fit the data and start test phase with test dataset.

1. Compile the model (`model.compile`)
2. Fit the data to the model (`model.fit`)
3. Evaluate model results (`model.evaluate`) using separate test data.

````python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D
from tensorflow.keras.losses import sparse_categorical_crossentropy
from tensorflow.keras.optimizers import Adam

# TODO: Load CIFAR-10 data with cifar10.load_data()
(input_train, target_train), (input_test, target_test) = cifar10.load_data()

# TODO: Let's scale the data after parsing
# Converting to float32 and scale to 0-1 range
input_train = input_train.astype('float32') / 255.0
target_train = target_train.astype('float32')
input_test = input_test.astype('float32') / 255.0
target_test = target_test.astype('float32')
#print(input_train.dtype)

# TODO: Create a deep learning model.
# TODO: Create a model with Keras
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', padding='same', strides=2, input_shape=(32, 32, 3)),
    Conv2D(64, (3, 3), activation='relu', padding='same', strides=1),
    Flatten(),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.4),
    Dense(10, activation='softmax')
])

# TODO: Compile the model
model.compile(optimizer=Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
# TODO: Set the data in the model
model.fit(input_train, target_train, epochs=5, batch_size=50)
````

````python
# Answers to the assignment.
# Note! Do not edit this cell, just run it after you complete the task.

# TODO: Create a summary of the model and print evaluation metrics
model.summary()
score = model.evaluate(input_test, target_test, verbose=0)
print(f'Loss: {score[0]} and accuracy: {score[1]}')
````

### RNN and LSTM

_Recurrent neural networks_ (**RNN**) are particularly well suited for sequential data. Well-known practical applications of RNN include e.g. Google's voice search and Apple's Siri.

* RNN has internal memory.
* RNN is a deep learning algorithm that can remember the input.
* Applications of RNN are **LSTM** and **GRU**
* RNN is the recommended algorithm for sequential data, e.g., for written text, time series, weather forecasts, economic data, etc.

In an RNN, the data circulates in an internal loop. When the RNN calculates the output layer, it also considers the previous layer and also the layer before it, because it has short-term memory.

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, LSTM 

# Read txt file line by line
with open('./data/clickbait_data.txt', 'r', encoding='utf-8') as clickbait_file:
    # Read all lines from the file and remove empty lines
    clickbait_data = [line.strip() for line in clickbait_file if line.strip()]
    
with open('./data/non_clickbait_data.txt', 'r', encoding='utf-8') as non_clickbait_file:
    # Read all lines from the file and remove empty lines
    non_clickbait_data = [line.strip() for line in non_clickbait_file if line.strip()]


# Create a DataFrame with one column (text) containing the lines
clickbait = pd.DataFrame({'text': clickbait_data})
non_clickbait = pd.DataFrame({'text': non_clickbait_data})

#clickbait.head()
#non_clickbait.head()

# Add class column and combine two dataframes
clickbait['class'] = 1
non_clickbait['class'] = 0
df = pd.concat([clickbait, non_clickbait], ignore_index=True)

print(f'Size of data: {len(df)}')
df.head()
````

````python
#  Split the data as Train test
X = df.drop(columns=['class'])
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2302)

tokenizer = Tokenizer()

# Data has text using .fit_on_texts
tokenizer.fit_on_texts(df['text'])
# Will be used in tokernizer as parameter
words_count = len(tokenizer.word_index)
print(f'Word count: {words_count}')

# Converting to list, tokenizer expects a list
X_train = X_train['text'].tolist()
X_test = X_test['text'].tolist()

# out_of_vocabulary parameter is used to assign token to words that were notin training. Ensures consistency
tokenizer_2= Tokenizer(num_words=words_count, oov_token="<OOV>")
tokenizer_2.fit_on_texts(X_train)

# Find longest texts total "word" length in data set. Used in padding
max_len = df['text'].apply(lambda x: len(x.split())).max()
print(f'Longest text: {max_len}')

# Converting texts to sequences and add padding
X_train = tokenizer_2.texts_to_sequences(X_train)
X_test = tokenizer_2.texts_to_sequences(X_test)

X_train = pad_sequences(X_train, maxlen=max_len)
X_test = pad_sequences(X_test, maxlen=max_len)
````

````python
# LSTM Model

model = Sequential([
    Embedding(words_count, 50), 
    LSTM(32),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='rmsprop', # Common ins RNNs
              loss='binary_crossentropy',# Loss commonly used in binary classification
              metrics=['accuracy'])

model.fit(X_train, y_train,
          epochs=10,
          batch_size=64,
          validation_split=0.2) 
````

````python
# Answers to the assignment.
# Note! Do not edit this cell, just run it after you complete the task. 

# TODO: Put the result of the model.evaluate() function in the results variable.
# TODO: set metrics=['accuracy'] to monitor in model.compile() to see how much of the neural network's predictions is correct at different stages of training.
results = model.evaluate(X_test, y_test)
print(f"Test Loss:{results[0]} Test Accuracy:{results[1]*100}%")
````

Model performed quite well. It is able to predict clickbait text from non-clickbait with 98% accuracy
