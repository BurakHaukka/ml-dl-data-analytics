## Introduction 

  ### Data Collection  

![Wood Tiger Moth](wtm.PNG)

Individual wood tiger moths collected between years 2009-2020. Samples are collected with traps and nets.  
Wood tiger moths have 4 wings, 2 forewing and 2 hindwing. Wood tiger moth coloration is white, yellow, orange and red. In addition to that they carry melanised dark areas, which is the main interest in the project.

After sample collection, photos were taken and processed to extract coloring information. ROI (regions of interest) are marked and processed to extract the information needed, such as wing area, coloring area (colored and melanised) and more, which will not be needed.

  ### Data Set

Data consists 1754 individual moths.

Features:
- Label: Unique label for each individual.
- total_wing_area: Sum of fore and hind wing aread.
- total_melanization_area: Area of only black color.
- prop: Proportion of melanisation. 
- year: Year of collection.
- Sex: Sex of the individual F/M.
- Color: Color identification.
- Melanization: Hindwing phenotype plus/minus.
- region: Area that sample was collected.
    - NFI: Northern Finland
    - CFI: Central Finland
    - SFI: Southern Finland
    - EST: Estonia
- ow_temp: Over-wintering temperature.
- ll_temp: Large larval stage temperature.
- ad_temp: Adult stage temperature.
- sl_temp: Small larval stage temperature.
- ow_hum: Over-wintering humidity.
- ll_hum: Large larval stage humidity.
- ad_hum: Adult stage humidity.
- sl_hum: Small larval stage humidity.
- ow_prec: Over-wintering precipitation.
- ll_prec: Large larval stage precipitation.
- ad_prec: Adult stage precipitation.
- sl_prec: Small larval stage precipitation.
- fore_prop: Forewing melanisation proportion.
- hind_prop: Hindwing melanisation proportion.

Weather data is collected from Finnish Meteorological Institute and Estonian Environment Agency.
Weather data that was provided are monthly averages of temperature, humidity and precipitation.

  ### Feature Engineering  

At the time I used **R-Statistics programming** language for my analysis. All the calculations are done through R-programming
During my data analysis, I modified some of the raw data to new features. I will explain how I created them here. Raw data contained areas in pixels.

- prop: Proportion of melanisation was calculated from total_melanization / total_wing_area
- fore_prop: forewing_melanisation / forewing_area
- hind_prop: hindwing_melanisation / hindwing_area

Life cycle of wood tiger moth is, **Small Larva -> Over-wintering -> Large Larva -> Adult** then new generation of small larvae starts.  
- Small Larva: August, September
- Over-wintering:  October, November, December, January, February, March
- Large Larva: April, May
- Adult: June, July

For further analysis created features related to life cycle. I calculated averages of the weather data from months corresponding to life cycle stages. For example, overwintering temperature **ow_temp** is average of sum of October, November, December, January, February, March average temperatures.

```python
# Divide data set to regions
nfi_data = data_lstm[data_lstm['region'] == 'NFI']
cfi_data = data_lstm[data_lstm['region'] == 'CFI']
sfi_data = data_lstm[data_lstm['region'] == 'SFI']
est_data = data_lstm[data_lstm['region'] == 'EST']

#Sorting the years in ascending order
nfi_data = nfi_data.sort_values(by='year')
cfi_data = cfi_data.sort_values(by='year')
sfi_data = sfi_data.sort_values(by='year')
est_data = est_data.sort_values(by='year')

# Move prop to last
column_to_move = 'prop'

# Move the column to the last position
nfi_data = pd.concat([nfi_data.drop(columns=[column_to_move]), nfi_data[column_to_move]], axis=1)
cfi_data = pd.concat([cfi_data.drop(columns=[column_to_move]), cfi_data[column_to_move]], axis=1)
sfi_data = pd.concat([sfi_data.drop(columns=[column_to_move]), sfi_data[column_to_move]], axis=1)
est_data = pd.concat([est_data.drop(columns=[column_to_move]), est_data[column_to_move]], axis=1)

# Drop regions since it is not necessary
nfi_data = nfi_data.drop(columns=['region'])
cfi_data = cfi_data.drop(columns=['region'])
sfi_data = sfi_data.drop(columns=['region'])
est_data = est_data.drop(columns=['region'])

#print(nfi_data.head())
#print(cfi_data.head())
#print(sfi_data.head())
#print(est_data.head())

nfi_len = len(nfi_data)
cfi_len = len(cfi_data)
sfi_len = len(sfi_data)
est_len = len(est_data)

print("Sample size NFI:", nfi_len)
print("Sample size CFI:", cfi_len)
print("Sample size SFI:", sfi_len)
print("Sample size EST:", est_len)

# Calculate mean proportion for each year
nfi_prop_year = nfi_data.groupby('year')['prop'].mean().reset_index()
cfi_prop_year = cfi_data.groupby('year')['prop'].mean().reset_index()
sfi_prop_year = sfi_data.groupby('year')['prop'].mean().reset_index()
est_prop_year = est_data.groupby('year')['prop'].mean().reset_index()

# Display the resulting means
#print("NFI yearly mean: \n", nfi_prop_year)
#print("\nCFI yearly mean: \n", cfi_prop_year)
#print("\nSFI yearly mean: \n", sfi_prop_year)
print("\nEST yearly mean: \n", est_prop_year)
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split

# Prepare data for LSTM
def create_sequences(nfi_prop_year, sequence_length):
    X, y = [], []
    for i in range(len(nfi_prop_year) - sequence_length):
        X.append(nfi_prop_year[i:i+sequence_length])
        y.append(nfi_prop_year[i+sequence_length])
    return np.array(X), np.array(y)

sequence_length = 10  # 10 years, predicting 2020
X, y = create_sequences(nfi_prop_year['prop'], sequence_length)

# Split the data into train and test sets, time sequence is maintained. shuffle = False
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Reshape input data for LSTM (samples, time steps, features)

print(X_train)
print("Shape of X_train:", X_train.shape)
print("X_train dtype:", X_train.dtype)
print("y_train dtype:", y_train.dtype)
```

```python
# Build the LSTM model
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    LSTM(units=50),
    Dense(units=1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model. Because data is too small, validation_split = 0.0
model.fit(X_train, y_train, epochs=500, batch_size=32, validation_split=0.0, verbose=1)

# Evaluate the model
test_loss = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}')

# Make predictions
predictions = model.predict(X_test)

# Inverse transform the predictions
print("Prediction: ", predictions)

# Predict future values
future_data = nfi_prop_year['prop'].values[-sequence_length:]  # Get the last sequence of known data
for _ in range(1):
    # Reshape future_data to match the input shape expected by the model
    future_data_reshaped = future_data.reshape(1, -1, 1)
    # Predict the next value
    next_value = model.predict(future_data_reshaped)
    # Append the predicted value to future_data
    future_data = np.append(future_data, next_value)
print(future_data)
```

```python
# Keep only Labels that contain f1 and h1

filtered_df = full_raw[full_raw['Label'].str.contains('_f1|_h1')]
sample_data = filtered_df.iloc[:2]
len_filtered = len(filtered_df)
print("Length of the Filtered data: ", len_filtered)
print(sample_data)
```

```python
# Filtered_df contains the DataFrame with 'f1' and 'h1' labels
# Extract identifiers from the labels
filtered_df['Identifier'] = filtered_df['Label'].str.split('_').str[0]

# Split the DataFrame into two based on 'f1' and 'h1' labels
f1_df = filtered_df[filtered_df['Label'].str.contains('_f1')]
h1_df = filtered_df[filtered_df['Label'].str.contains('_h1')]

# Merge the two DataFrames based on the Identifier column
merged_df = pd.merge(f1_df, h1_df, on='Identifier', suffixes=('_f1', '_h1'))

# Drop unnecessary columns after merging, like 'Label_f1' and 'Label_h1'
merged_df.reset_index(drop=True, inplace=True)

# Rename the 'Label_f1' column to 'Label'
merged_df.rename(columns={'Label_f1': 'Label'}, inplace=True)

# Remove the '_f1' suffix from the 'Label' column
merged_df['Label'] = merged_df['Label'].str.replace('_f1', '')

missing_values_count = merged_df.isna().sum()
print("Missing data in merged dataframe: ", missing_values_count)

# Very little missing data, dropping the NaN values
merged_df.dropna(inplace=True)

# Drop extra Label_h1 column and Identifier
merged_df = merged_df.drop(columns=['Label_h1', 'Identifier'])

missing_values_count = merged_df.isna().sum()
print("Missing data in merged dataframe: ", missing_values_count)

merged_len = len(merged_df)
print("\nLength of modified data: ", merged_len)

sample_data = merged_df.iloc[:2]
print("\nSample data: \n",sample_data)
```

```python
# Need to add "region" feature for classification target

# A function to determine the region based on the starting letters in the label
def get_region(label):
    if label.startswith(('CFI', 'KS')):
        return 'CFI'
    elif label.startswith(('EST', 'V')):
        return 'EST'
    elif label.startswith(('N', 'NF', 'NFI')):
        return 'NFI'
    elif label.startswith(('SFI', 'HAN', 'TVA')):
        return 'SFI'
    else:
        return None

# Create a new 'region' column
merged_df['region'] = merged_df['Label'].apply(get_region)

sample_data = merged_df.iloc[1]
print("\nSample data: \n",sample_data)

missing_values_count = merged_df.isna().sum()
print("\nMissing data in merged dataframe:\n", missing_values_count)

# There are regions that are not related to this analysis. Dropping NaN value region columns
merged_df.dropna(inplace=True)
missing_values_count = merged_df.isna().sum()
print("\nMissing data in merged dataframe after drop:\n", missing_values_count)

merged_len = len(merged_df)
print("\nLength of modified data\n", merged_len)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix

# X contains the features and y contains the target (region)
X = merged_df.drop(columns=['Label', 'region'])  # 'Label' is not a feature
y = merged_df['region']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2302)

# Scale the features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train the MLPClassifier (Multi-layer Perceptron Classifier)
# Tried 300, 100, 50 -> 54%
clf = MLPClassifier(hidden_layer_sizes=(300, 100, 50, 25), max_iter=500, random_state=2302)
clf.fit(X_train_scaled, y_train)

# Evaluate the model
train_score = clf.score(X_train_scaled, y_train)
test_score = clf.score(X_test_scaled, y_test)

print(f"Training Accuracy: {train_score:.2f}")
print(f"Test Accuracy: {test_score:.2f}")

y_pred = clf.predict(X_test_scaled)

# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)
#print(cm)
# Get unique class labels
classes = np.unique(y_test)

# Plot confusion matrix
plt.figure(figsize=(6, 4))
sns.set(font_scale=1.2)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Testing with thesis data for comparison
# Prepare data: X (features) and y (target)
X2 = data.drop(columns=['region'])
y2 = data['region']

# Split the data into train and test sets
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.33, random_state=2302)

# Initialize and train the MLPClassifier (Multi-layer Perceptron Classifier)
# Tried 300, 200, 100 -> 36%
clf2 = MLPClassifier(hidden_layer_sizes=(300, 200, 100, 50), max_iter=500, random_state=2302)
clf2.fit(X2_train, y2_train)

# Evaluate the model
train_score2 = clf2.score(X2_train, y2_train)
test_score2 = clf2.score(X2_test, y2_test)

print(f"Training Accuracy: {train_score2:.2f}")
print(f"Test Accuracy: {test_score2:.2f}")

y_pred2 = clf2.predict(X2_test)

# Calculate confusion matrix
cm2 = confusion_matrix(y2_test, y_pred2)
#print(cm2)
# Get unique class labels
classes = np.unique(y2_test)

# Plot confusion matrix
plt.figure(figsize=(6, 4))
sns.set(font_scale=1.2)
sns.heatmap(cm2, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

## Select a Specific Year

I want to see how classification will work if I selected specific year.  

But labels of some years are not in correct format. This makes accurate extraction harder.  Raw image data itself does not contain years, but the year is coded in the Label.  
Label structure is approximately region-year-number. In some cases region-year-letter-number.  
Wrong labeling includes regionyear-number. 

```python
import re

merged_year_df = merged_df
def extract_year(label):
    # Find the first sequence of digits in the label
    match = re.search(r'\b(\d{2})\b', label)
    if match:
        try:
            year = int(match.group(1))  # Extract the first two digits as the potential year
            # Check if the integer is within the range '09' to '20'
            if year >= 9 and year <= 20:
                return year + 2000
        except ValueError:
            pass  # Ignore if conversion to integer fails
    return None  # Return None if no valid year is found or conversion fails

# Apply the function to extract year and create a new 'year' feature
merged_year_df['year'] = merged_year_df['Label'].apply(extract_year).astype('Int32')

# Display the updated DataFrame with the 'year' feature
sample_data2 = merged_year_df.iloc[:2]
print(sample_data2)

region_counts = merged_df['region'].value_counts()
print("\nThesis data region counts:\n", region_counts)

# Count the occurrences of each year in the 'year' column
year_counts = merged_df['year'].value_counts()

# Display the counts for each year
print("\nYear extraction rows:\n", year_counts)

thesis_counts = full_data['year'].value_counts()
print("\nThesis years rows:\n", thesis_counts)
```

## Discussion  
From now on data used in the models will be referred as **thesis data** and **raw image data**.

### Data

![Wood Tiger Moth](wtm.PNG)  
**Thesis data** is built for a specific purpose. It includes regional weather information and melanisation proportion.  
Overall the data set does not seem to be good for predicting the region.  
There are two main reasons for this:
 - Data does not have enough distinguishing features for individuals.
 - **prop**, **hind_prop**, **fore_prop** features have overlapping distribution between populations. Meaning, there is no clear distinction between populations in terms of melanin coverage. This leads to miss labeling of individuals.

**Raw image data** is different from thesis data. This data contains information extracted from images of moths and contains more distinguishing information.  

### Models

Using **thesis data** with MLP Classifier model, I was able to get **38%** accuracy when I dropped **prop** feature. This feature is actually redundant, since it is over-all melanisation. **hind_prop** and **fore_prop** are more specific for each region and increased the accuracy by few percents.

Used **raw image data** with MLP Classifier model. Data only contains **left side fore and hind wings** information. This is justifiable by assuming wings are symmetrical. Just by using two wings, model was able to predict **region** with **57%** accuracy. This is most likely due to the fact that, data set gives better view on distinguishing features on color information for different regions.
